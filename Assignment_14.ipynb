2. From the Pillow module, how do you get the RGBA value of any images?

__Ans:-__

```python
from PIL import Image
img = Image.open('space.jpg')
img
```

```python
data = img.convert('RGBA').getdata()
```

```python
counter = 0
# Just take first 10 RGBA values
while counter < 10:
    print(data[counter])
    counter += 1
```
